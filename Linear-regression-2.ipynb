```python
x= df[["TV", "radio", "newspaper"]]
x
```

```python
y= df["sales"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state = 42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
model.coef_
```

```python
predictions
```

```python
result = pd.DataFrame({"Actual": y_test,"Predicted": predictions, "Difference": y_test - predictions})
result
```

```python
error = np.abs(predictions- y_test)
plt.scatter(predictions, y_test, c= error, cmap = "coolwarm", s= 50)
plt.xlabel("Predicted values")
plt.ylabel("Actual values")
plt.ylabel("Actual prediction vs actual values")
plt.colorbar(label ="Error")
plt.show()
```
